```python
import os, kagglehub, subprocess
```

```python
ollama_file_path = '../ollama_file'
os.makedirs(ollama_file_path, exist_ok=True)
abs_folder_path = os.path.abspath(ollama_file_path)
abs_folder_path
```

```python
print("Archiving Ollama Model Cache...")

# 1. Ollama가 생성한 데이터 폴더 (.ollama)를 통째로 압축
# 우리는 아까 HOME=/tmp로 설정했으니, 데이터는 /tmp/.ollama 에 있습니다.
# 압축 파일은 /kaggle/working (Output)에 저장됩니다.
subprocess.run(f"tar -czf {abs_folder_path}/gpt_oss_120b_ollama.tar.gz -C /root .ollama", shell=True)

print(f"Archive created at {abs_folder_path}/gpt_oss_120b_ollama.tar.gz")
print("이제 우측 패널 Output에서 이 파일을 다운로드 받아서, 새 데이터셋으로 올리세요!")
```
